## MultiQC Report
Overall, the RNA-seq data were of high quality across the three control replicates, with total read counts ranging from ~84 to ~119 million reads per sample. Alignment efficiency was excellent, with over 97% of reads mapped and ~93–94% of reads uniquely aligned to the reference genome, indicating low levels of contamination or technical artifacts. The average read length was 100–200 bp, and mismatch and indel rates remained below 0.5%, consistent with accurate sequencing. FastQC analysis showed that per-base sequence quality scores remained high (median Phred >30 across all cycles), suggesting low error rates. While a small fraction of reads contained overrepresented adapter sequences (≤0.02% of total reads), this level is not expected to substantially bias downstream analyses. Duplication levels were modest (~9–18%), which is acceptable for RNA-seq libraries and likely reflects high sequencing depth rather than severe library complexity issues. Collectively, these results indicate that the dataset is of sufficient quality for reliable downstream differential expression analysis, with only minor adapter and duplication signals warranting caution in interpretation.
